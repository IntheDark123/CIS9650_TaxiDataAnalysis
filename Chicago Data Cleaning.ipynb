```python
dfc.dropna(subset = ['Trip End Timestamp'], inplace = True)
# Drop null end times as it causes problems during conversion to datetime datatype
```

Tasks:
1. Need to change timestamp fields from object to datetime
2. Adding date time columns such as day of month, year, month, day of week

```python
dfc['Trip Start Timestamp'] = pd.to_datetime(dfc['Trip Start Timestamp'])
                                             #, format="%m/%d/%Y %I:%M:%S %p")
```

```python
dfc['Trip End Timestamp'] = pd.to_datetime(dfc['Trip End Timestamp'])
                                           #, format="%m/%d/%Y %I:%M:%S %p")
```

```python
dfc.rename(columns={'Trip Start Timestamp':'TripStart','Trip End Timestamp':'TripEnd'}, inplace = True)
```

```python
dfc.drop(['Trip ID','Taxi ID'], axis = 1, inplace = True)
dfc.rename(columns={'Trip Start Timestamp':'TripStart','Trip End Timestamp':'TripEnd'}, inplace = True)
```

```python
dfc.head()
```

```python
type(dfc['TripStart'])
```

```python
dateFeats = ['TripStart','TripEnd']
```

```python
for feat in dateFeats:
    dfc[feat+'_dayofmonth'] = dfc[feat].dt.day
    dfc[feat+'_month'] = dfc[feat].dt.month
    dfc[feat+'_year'] = dfc[feat].dt.year
    dfc[feat+'_dayofweek'] = dfc[feat].dt.dayofweek
    dfc[feat+'_week'] = dfc[feat].dt.week
    dfc[feat+'_time'] = dfc[feat].dt.time
    dfc[feat+'_hourofday'] = dfc[feat].dt.hour
    dfc[feat+'_minuteofday'] = dfc[feat].dt.minute
```
